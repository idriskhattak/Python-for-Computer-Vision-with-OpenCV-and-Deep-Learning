```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Let's create a blank image

```python
blank_img = np.zeros(shape=(500,500,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

Creting a rectangle

```python
cv2.rectangle(blank_img, pt1=(300,100), pt2=(500,50), color=(0,255,0), thickness=5)
```

```python
plt.imshow(blank_img)
```

Creating a square

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,211,0),thickness=5)
plt.imshow(blank_img)
```

Creating a circle

```python
cv2.circle(blank_img,center=(100,100),radius=(50),color=(0,150,0),thickness=5)
plt.imshow(blank_img)
```

Creating a circle that it's inside is all colored same method goes for all, if want color all of the inside of any shape

```python
cv2.circle(blank_img,center=(400,400),radius=(50),color=(0,150,0),thickness=-1)
plt.imshow(blank_img)
```

Creating a line

```python
cv2.line(blank_img,pt1=(0,0),pt2=(500,500),color=(255,0,0),thickness=3)
plt.imshow(blank_img)
```
