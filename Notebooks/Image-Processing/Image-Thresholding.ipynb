```python
import cv2
import matplotlib.pyplot as plt
```

Creating a function that load an image as greyscale

```python
def load(path):
    img = cv2.imread(path,0)  # 0 is used to load an image in greyscale
    if img is None:
        print('Error loading image')
    else:
       return img
```

```python
path = 'E:\\Python-for-Computer-Vision-with-OpenCV-and-Deep-Learning\\DATA\\rainbow.jpg'
```

```python
img1 = load(path)
```

```python
plt.imshow(img1,cmap='grey')
```

```python
thresh_1_val,thresh_1_img = cv2.threshold(img1,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh_1_img,cmap='grey')
```

#   Lets work with real life example

```python
path = 'E:\\Python-for-Computer-Vision-with-OpenCV-and-Deep-Learning\\DATA\\crossword.jpg'
```

```python
img2 = load(path)
```

```python
plt.imshow(img2,cmap='grey')
```

Creating another function to show the image properly

```python
def show_img(img):
    fig = plt.figure(figsize=(12,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='grey')
    
```

```python
show_img(img2)
```

Now let's apply threshold to this image

```python
thresh_2_val,thresh_2_img = cv2.threshold(img2,180,255,cv2.THRESH_BINARY)
```

```python
show_img(thresh_2_img)
```

Using adaptive thresholding

```python
adap_thresh = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
```

```python
show_img(adap_thresh)
```

let's blend them together and see if we can get better results

```python
blended = cv2.addWeighted(src1=thresh_2_img,alpha=0.6,src2=adap_thresh,beta=0.4,gamma=0)
```

```python
show_img(blended)
```
